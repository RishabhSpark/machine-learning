```python
cat_cols = ['address', 'famsize', 'reason', 'schoolsup', 'famsup', 'paid', 'activities', 'higher', 'internet', 'romantic']
```

```python
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
X_new = pd.get_dummies(X, columns = cat_cols, drop_first = True)
X_new
```

```python
from sklearn.preprocessing import minmax_scale

min_max_cols = ['studytime', 'failures', 'freetime', 'goout', 'health', 'absences', 'G1', 'G2']

for i in min_max_cols:
    X_new[i] = minmax_scale(X_new[i])

X_new.drop(min_max_cols, axis=1, inplace=True)
X_new.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression


def splitting(t_size):
    X_train, X_test, y_train, y_test = train_test_split(X_new, Y, test_size=t_size, random_state=0)
    
    
    model = LinearRegression()
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)

    print('mean_squared_error : ', mean_squared_error(y_test, predictions))
    print('mean_absolute_error : ', r2_score(y_test, predictions))
       
```

```python
split_size_70_30 = splitting(0.3)
```

```python
split_size_80_20 = splitting(0.2)
```
