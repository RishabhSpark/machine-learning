```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("winequality-red.csv", delimiter = ";")
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['quality'] = df['quality'].astype(int)
df.head()
```

```python
df.loc[df['quality'] >=8, 'q'] = 3
df.loc[(df['quality'] <8) & (df['quality'] >=6) , 'q'] = 2
df.loc[df['quality'] <=5, 'q'] = 1
df.drop('quality', axis=1, inplace=True)
df.head()
```

```python
df.rename(columns = {'q':'quality'}, inplace = True)
df['quality'] = df['quality'].astype(int)
df.head()
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
Y = df.iloc[:,-1]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import recall_score, accuracy_score, confusion_matrix

def naives_bayes_model(t_size):
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=t_size, random_state=0)
    
    model = GaussianNB()
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print ("Accuracy:", accuracy)
    
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix: \n", cm)
    
    # recall = recall_score(y_test, y_pred)
    # print('Recall:', recall)
```

```python
naives_bayes_model(0.3)
```

```python
naives_bayes_model(0.2)
```
