```python
X = df.iloc[:,:-1]
X.shape
```

```python
Y = df.iloc[:,-1]
Y.shape
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split

sc = StandardScaler()
X_arr = sc.fit_transform(X)
le = LabelEncoder()
y = le.fit_transform(Y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train_df = pd.DataFrame(X_train, columns=X.columns)
X_train_df.head()
```

```python
from sklearn.decomposition import PCA

pca= PCA(n_components= 2)
lda = LinearDiscriminantAnalysis()
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)
```

```python
pca.explained_variance_ratio_
```

```python
plt.scatter(X_train_pca[:,0],X_train_pca[:,1],c=y_train,cmap='rainbow',alpha=0.7)
plt.title('PCA')
```

```python
plt.scatter(X_train_lda[:,0],y_train,c=y_train,cmap='rainbow',alpha=0.7)
plt.title('LDA')
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)
  
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(X_train_pca)
    kmeanModel.fit(X)
  
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / X.shape[0])
    inertias.append(kmeanModel.inertia_)
  
    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / X.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

```python
for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
classifier= KMeans(n_clusters= 2)
classifier.fit(X)
```
